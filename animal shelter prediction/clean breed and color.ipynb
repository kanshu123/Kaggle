```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv('train.csv')
bc = pd.DataFrame()
bc = train[['Breed', 'Color']]
bc['Mix'] = 'Pure'
breeds = bc["Breed"].apply(lambda x: x.split(" Mix")[0])
breeds = breeds.apply(lambda x: re.sub('Black\s?|Tan\s?', '', x))
breeds = breeds.apply(lambda x: re.sub('^/', '', x))
breeds = breeds.str.replace("//", "")
breeds = breeds.apply(lambda x: pd.Series(x.split("/")))
breeds.columns = ['Breed', 'SecondaryBreed']
breed_counts = breeds["Breed"].value_counts()
frequent_breeds = breed_counts[breed_counts > 50].to_dict().keys()
breeds[~breeds['Breed'].isin(frequent_breeds)] = "Rare"

bc.drop('Breed', inplace=True, axis = 1)
bc = pd.concat([bc, breeds], axis = 1)
bc.drop('SecondaryBreed', inplace=True,axis = 1)

bc.loc[train.Breed.str.contains('Mix'),'Mix'] = 'Mix'
bc.loc[train.Breed.str.contains('/'),'Mix'] = 'Mix'
```

```python
bc['Color'] = [x[0] for x in bc['Color'].str.split('/').tolist()]
```

```python
b_encoder = LabelEncoder()
c_encoder = LabelEncoder()
b_encoder = b_encoder.fit(bc.Breed)
c_encoder = c_encoder.fit(bc.Color)
bc.Breed = b_encoder.transform(bc.Breed)
bc.Color = c_encoder.transform(bc.Color)
```

```python
bc.drop('Mix', inplace = True,axis = 1)
```

```python
cleaned_train = pd.read_csv('cleaned_train.csv')
cleaned_train = pd.concat([cleaned_train,bc], axis = 1)
```

```python
cleaned_train.drop(['ColorCode','BreedCode'],inplace = True, axis = 1)
cleaned_train.to_csv('clean_train.csv')
```
