```python
import pandas as pd
import numpy as np
```

```python
act_train = pd.read_csv('act_train.csv')
act_test = pd.read_csv('act_test.csv')
people = pd.read_csv('people.csv')
sub_id = act_test['activity_id']
```

```python
act_train.columns
```

```python
train = pd.merge(act_train,people, how = 'left',on ='people_id',left_index = True)
test = pd.merge(act_test,people, how = 'left',on ='people_id',left_index = True)
```

```python
train_col = train.columns.tolist()
```

```python
train_col.remove('outcome')
```

```python
train_col
```

```python
train[train_col]
```

```python
from sklearn.feature_extraction import FeatureHasher
h = FeatureHasher(input_type = 'string')
data = h.fit_transform(train[train_col])
```

```python
train['outcome'] = train['outcome'].astype('str')
target = h.transform(train['outcome'])
```

```python
#np.set_printoptions(threshold=np.inf)
test = h.transform(test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
seed = 7
np.random.seed(seed)
```

```python
from keras.layers import advanced_activations
model = Sequential()
model.add(Dense(50,input_dim =1048576,init ='uniform',activation = 'relu' ))
model.add(advanced_activations.LeakyReLU(alpha=0.1))
#model.add(Dense(30,init = 'uniform',activation = 'relu'))
#model.add(Dense(20,init = 'uniform',activation = 'relu'))
#model.add(Dense(10,init = 'uniform',activation = 'relu'))
#model.add(Dense(10,init = 'uniform',activation = 'relu'))
#model.add(Dense(10,init = 'uniform',activation = 'relu'))
#model.add(Dense(10,init = 'uniform',activation = 'relu'))
model.add(Dense(1, init = 'uniform', activation = 'sigmoid'))
model.compile(loss ='binary_crossentropy',optimizer = 'sgd',metrics = ['accuracy'])
```

```python
model.fit(data,target,nb_epoch = 5,batch_size = 100,verbose = 2, shuffle = False,validation_split = 0.1)
```
